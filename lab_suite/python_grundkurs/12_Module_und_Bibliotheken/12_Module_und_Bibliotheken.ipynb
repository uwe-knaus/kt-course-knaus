## C: `#include` – ist das vergleichbar mit Python-Import?

In **C** bindest du Header und Bibliotheken so ein:

- **`#include <stdio.h>`** – Sucht die Datei in **System-/Compiler-Pfaden** (z. B. Standardbibliothek).
- **`#include "mein_header.h"`** – Sucht die Datei **relativ zum Projekt** oder in angegebenen Pfaden.

Der **Präprozessor** kopiert den Inhalt der Datei buchstäblich in deinen Quelltext. Es gibt **keine** Namensräume: Alle Makros und Deklarationen landen im gleichen globalen Raum; Namenskonflikte sind möglich.

**Python** macht es anders: **`import modul`** lädt Code zur **Laufzeit** und legt ihn in einem **eigenen Namensraum** ab (z. B. `math.sqrt`). Es wird nichts „reinkopiert"; du greifst gezielt auf `modul.name` zu. Insofern ist **`import`** nur grob vergleichbar mit `#include`: Beide holen „fremden" Code, aber die Mechanik (Laufzeit vs. Präprozessor, Namensräume) ist unterschiedlich.

## Was sind Module und Bibliotheken in Python?

- **Modul:** Eine **Python-Datei** (`.py`) mit Funktionen, Klassen und Variablen. Der Dateiname (ohne `.py`) ist der Modulname. Beispiel: `math.py` → Modul **`math`**.
- **Bibliothek / Paket:** Oft synonym verwendet. Genauer: Ein **Paket** ist ein Ordner mit Modulen und einer `__init__.py` (oder ab Python 3.3 auch ohne); eine **Bibliothek** meint meist ein **installierbares Paket** (z. B. von PyPI), das du mit **pip** installierst.

**Eingebaute Bibliothek (Standard Library):** Kommt mit Python mit – z. B. `math`, `os`, `sys`, `json`. Du musst nichts extra installieren.

**Externe Bibliotheken:** z. B. `numpy`, `matplotlib`. Du installierst sie mit **pip** (oft in einer **virtuellen Umgebung** `.venv`).

## Import: So bindest du Module ein

| Schreibweise | Bedeutung |
|--------------|----------|
| `import math` | Modul `math` laden; Nutzung z. B. **`math.sqrt(4)`** |
| `from math import sqrt, log2` | Nur `sqrt` und `log2` in den aktuellen Namensraum holen; Nutzung: **`sqrt(4)`**, **`log2(2)`** |
| `from math import *` | Alle Namen aus `math` importieren (ungünstig: Namenskonflikte möglich) |

Üblich und lesbar: **`import math`** oder **`from math import log2`**, wenn du nur wenige Funktionen brauchst.

```python
import math

print(math.sqrt(16))
print(math.log2(8))
print(math.pi)
```

```python
from math import log2, sqrt

print(sqrt(16))
print(log2(8))
```

## Eingebaute Bibliotheken (Standard Library)

Python bringt viele **Module** mit, die du ohne Installation nutzen kannst:

| Modul | Typischer Einsatz |
|-------|-------------------|
| **`math`** | `sqrt`, `log2`, `sin`, `pi`, … |
| **`os`** | Pfade, Umgebungsvariablen, Verzeichnisoperationen |
| **`sys`** | `sys.argv`, `sys.path`, `sys.version` |
| **`json`** | JSON lesen/schreiben |
| **`random`** | Zufallszahlen (gleichverteilt, Auswahl) |

Du brauchst dafür **kein** pip – sie gehören zur Python-Installation.

```python
import sys
import os

print("Python-Version:", sys.version.split()[0])
print("Aktuelles Verzeichnis:", os.getcwd())
```

## pip: Pakete aus dem Internet installieren

**pip** ist der Standard-**Paketmanager** für Python. Er lädt Pakete von **PyPI** (Python Package Index) und installiert sie.

- **Installation eines Pakets:** `pip install numpy`
- **Mehrere Pakete:** `pip install numpy matplotlib`
- **Aus einer Datei:** `pip install -r requirements.txt` (z. B. im KT-workspace)

**Wichtig:** Ohne virtuelle Umgebung installiert pip **systemweit** (oder benutzerweit). Für Projekte ist das ungünstig: Verschiedene Projekte brauchen unterschiedliche Versionen. Deshalb nutzt man eine **virtuelle Umgebung** (`.venv`).

## Warum wir `.venv` verwenden

Eine **virtuelle Umgebung** (oft im Ordner **`.venv`**) ist ein **eigenes Python-Verzeichnis** nur für dein Projekt:

1. **Isolation:** Pakete, die du mit `pip install` einrichtest, landen **nur** in dieser Umgebung – nicht systemweit.
2. **Reproduzierbarkeit:** Du kannst mit **`pip freeze > requirements.txt`** die genutzten Versionen speichern. Andere (oder du auf einem anderen Rechner) installieren mit **`pip install -r requirements.txt`** exakt dieselben Versionen.
3. **Keine Konflikte:** Verschiedene Projekte können unterschiedliche Versionen von z. B. `numpy` nutzen.

**Aktivierung (Beispiele):**
- Windows (PowerShell): **`.venv\\Scripts\\Activate.ps1`**
- Linux/macOS: **`source .venv/bin/activate`**

Nach der Aktivierung zeigt `which python` / `where python` auf das Python in `.venv`; alle `pip install`-Befehle betreffen nur diese Umgebung.

## Beispiele: Standardbibliothek und externe Pakete

**math** (eingebaut) – z. B. für Informationstheorie (Entropie, Logarithmen):

```python
from math import log2

p = 0.5
h = -p * log2(p) - (1 - p) * log2(1 - p)
print("Entropie H(p) für p = 0.5:", round(h, 4))
```

**numpy** und **matplotlib** (extern, mit pip installierbar) – du hast sie bereits in früheren Kapiteln genutzt:

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x))
plt.xlabel("x")
plt.title("Sinuskurve (numpy + matplotlib)")
plt.grid(True)
plt.show()
```

## Kurz zusammengefasst

| Thema | Inhalt |
|-------|--------|
| **C vs. Python** | C: `#include <…>` / `"…"` kopiert Header; Python: `import` lädt zur Laufzeit mit eigenem Namensraum. |
| **Modul** | Eine `.py`-Datei; Name = Dateiname ohne `.py`. |
| **Bibliothek/Paket** | Sammlung von Modulen; externe z. B. von PyPI. |
| **Standard Library** | Mit Python mitgeliefert: `math`, `os`, `sys`, `json`, `random`, … |
| **pip** | Installiert Pakete von PyPI; z. B. `pip install numpy`. |
| **.venv** | Virtuelle Umgebung: isolierte Pakete pro Projekt, reproduzierbar mit `requirements.txt`. |

Damit kannst du gezielt Module einbinden – eingebaute ohne pip, externe mit pip in einer `.venv`.
