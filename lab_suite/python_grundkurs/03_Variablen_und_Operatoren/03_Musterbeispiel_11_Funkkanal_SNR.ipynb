## Fragestellung

Wir haben einen **Funkkanal mit 8 MHz Bandbreite** zur Verfügung. Welches **SNR** (Signal/Störleistungsverhaeltnis) brauchen wir, um ein digitalisiertes **Videosignal** (Abtastfrequenz 10 MHz, 8 Bit/Sample) uebertragen zu können?

## Antwort (aus dem Skriptum)

- Erforderliche Kapazität: $10\,\mathrm{MHz} \cdot 8\,\mathrm{bit} = 80\,\mathrm{Mbit/s}$.

- Shannon: $C = B \cdot \mathrm{ld}(1+\mathrm{SNR})$. Mit $C=80\,\mathrm{Mbit/s}$, $B=8\,\mathrm{MHz}$:

$$\mathrm{ld}(1+\mathrm{SNR}) = \frac{C}{B} = 10 \quad \Rightarrow \quad 1+\mathrm{SNR} = 2^{10} = 1024 \quad \Rightarrow \quad \mathrm{SNR} = 1023 \approx 30{,}1\,\mathrm{dB}$$
