## Fragestellung

Ein Audiorecorder nimmt ein **Stereo-Audiosignal** mit einer **Abtastrate von 44,1 kHz** und einer **Amplitudenauflösung von 16 bit** auf. Wieviel **Speicherplatz** ist für **1 Stunde** notwendig, wenn keine Datenkompression gemacht wird (z. B. WAV-Datei)?

## Antwort (aus dem Skriptum)

- Pro Sekunde: $44\,100$ Samples, 16 Bit pro Sample, **2 Kanäle** (Stereo):
  $$44\,100 \cdot 16 \cdot 2 = 705\,600\,\mathrm{bit/s}$$

- 1 Stunde = 3600 s:
  $$705\,600 \cdot 3600\,\mathrm{bit} = 2\,540\,160\,000\,\mathrm{bit} = 317\,520\,000\,\mathrm{Byte} \approx 302{,}8\,\mathrm{MByte}$$
