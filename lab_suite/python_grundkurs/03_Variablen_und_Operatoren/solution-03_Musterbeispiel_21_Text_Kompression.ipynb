## Fragestellung

Ein Text mit **80 000 Zeichen** (32 verschiedene Zeichen = Zeichenvorrat) wird codiert; der mittlere Binärstellenaufwand beträgt $\bar{m} = 2{,}2$ bit/Zeichen.

**a)** Wieviel Speicherplatz in **Bytes** ist bei idealer Kompression notwendig?

**b)** Welche **Bandbreite** muss ein Übertragungskanal bei **SNR = 20 dB** haben, damit der komprimierte Text in **20 s** übertragen werden kann?

**c)** Wie lange dauert die Übertragung über diesen Kanal, wenn der Text **nicht** komprimiert wird?

## Antwort (aus dem Skriptum)

**a)** $80\,000 \cdot 2{,}2 = 176\,000$ Bit $= 22\,000$ Byte.

**b)** Erforderliche Kapazität $C = 176\,000 / 20 = 8800\,\mathrm{bit/s}$. Mit $C = B \cdot \mathrm{ld}(1+\mathrm{SNR})$, SNR 20 dB $\Rightarrow$ 100: $8800 = B \cdot \mathrm{ld}(101) \approx B \cdot 6{,}66$ $\Rightarrow$ $B \approx 1322\,\mathrm{Hz}$.

**c)** Unkomprimiert: $\mathrm{ld}(32)=5$ bit/Zeichen. Verhältnis $5/2{,}2$; Dauer $T = 20 \cdot (5/2{,}2) \approx 45{,}45\,\mathrm{s}$.

## Python-Code

```python
import math

zeichen = 80_000
m_quer = 2.2
t_sec = 20
snr_db = 20
zeichenvorrat = 32

bits_komprimiert = zeichen * m_quer
bytes_komprimiert = int(bits_komprimiert // 8)
print("a) Speicher:", bytes_komprimiert, "Byte")

C = bits_komprimiert / t_sec
snr_linear = 10 ** (snr_db / 10)
ld_term = math.log2(1 + snr_linear)
B = C / ld_term
print("b) C =", C, "bit/s, B ca.", round(B, 0), "Hz")

bit_pro_zeichen_unkomp = math.log2(zeichenvorrat)
T_unkomp = t_sec * (bit_pro_zeichen_unkomp / m_quer)
print("c) Unkomprimiert:", bit_pro_zeichen_unkomp, "bit/Zeichen => T =", round(T_unkomp, 2), "s")
```
