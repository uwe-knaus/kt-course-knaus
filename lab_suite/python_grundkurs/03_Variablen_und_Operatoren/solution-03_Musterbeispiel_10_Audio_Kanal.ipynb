## Fragestellung

**a)** Ein Audiosignal mit einer Bandbreite von **300 Hz bis 3300 Hz** soll mit **8 bit/Sample** abgetastet und übertragen werden. Welche **Kanalkapazität** in bit/s ist dafür notwendig?

**b)** Es steht ein Übertragungskanal mit einer Bandbreite von **10 kHz** und einem Signal/Störleistungsverhältnis **SNR = 20 dB** zur Verfügung. Kann das Audiosignal aus a) über diesen Kanal übertragen werden?

## Antwort (aus dem Skriptum)

**a)** Nyquist: minimale Abtastrate $\geq 2 \cdot f_\max = 2 \cdot 3300 = 6600\,\mathrm{Hz}$.

$$C = 6600\,\mathrm{Hz} \cdot 8\,\mathrm{bit} = 52\,800\,\mathrm{bit/s} = 52{,}8\,\mathrm{kbit/s}$$

**b)** SNR 20 dB $\Rightarrow$ lineares Verhältnis $10^{20/10} = 100$. Shannon:

$$C = B \cdot \mathrm{ld}(1+\mathrm{SNR}) = 10\,000 \cdot \mathrm{ld}(101) \approx 66{,}58\,\mathrm{kbit/s}$$

Der Kanal hat ausreichende Kapazität (66,58 > 52,8 kbit/s).

## Python-Code

```python
import math

f_max = 3300
bit_pro_sample = 8
f_abtast_min = 2 * f_max
C_erforderlich = f_abtast_min * bit_pro_sample
print("a) Abtastrate min.", f_abtast_min, "Hz => C =", C_erforderlich, "bit/s")

B = 10_000
snr_db = 20
snr_linear = 10 ** (snr_db / 10)
C_kanal = B * math.log2(1 + snr_linear)
print("b) Kanal C =", round(C_kanal, 2), "bit/s")
print("   C_kanal > C_erforderlich:", C_kanal > C_erforderlich)
```
