## Fragestellung

Wir tasten ein Signal mit einer **Taktfrequenz von 10 kHz** ab. Welche **Kanalkapazität** brauchen wir für die Übertragung, wenn

a) **1024** Quantisierungsstufen,

b) **16 384** Quantisierungsstufen

bei der Abtastung verwendet werden?

## Antwort (aus dem Skriptum)

- **a)** $1024 = 2^{10}$ → **10 Bit pro Sample**.
  $$C = 10\,\mathrm{bit} \cdot 10\,000\,\mathrm{Hz} = 100\,\mathrm{kbit/s}$$

- **b)** $16\,384 = 2^{14}$ → **14 Bit pro Sample**.
  $$C = 14\,\mathrm{bit} \cdot 10\,000\,\mathrm{Hz} = 140\,\mathrm{kbit/s}$$
