## Fragestellung

Ein Text mit **80 000 Zeichen** (32 verschiedene Zeichen = Zeichenvorrat) wird codiert; der mittlere Binärstellenaufwand beträgt $\bar{m} = 2{,}2$ bit/Zeichen.

**a)** Wieviel Speicherplatz in **Bytes** ist bei idealer Kompression notwendig?

**b)** Welche **Bandbreite** muss ein Übertragungskanal bei **SNR = 20 dB** haben, damit der komprimierte Text in **20 s** uebertragen werden kann?

**c)** Wie lange dauert die Übertragung ueber diesen Kanal, wenn der Text **nicht** komprimiert wird?

## Antwort (aus dem Skriptum)

**a)** $80\,000 \cdot 2{,}2 = 176\,000$ Bit $= 22\,000$ Byte.

**b)** Erforderliche Kapazität $C = 176\,000 / 20 = 8800\,\mathrm{bit/s}$. Mit $C = B \cdot \mathrm{ld}(1+\mathrm{SNR})$, SNR 20 dB $\Rightarrow$ 100: $8800 = B \cdot \mathrm{ld}(101) \approx B \cdot 6{,}66$ $\Rightarrow$ $B \approx 1322\,\mathrm{Hz}$.

**c)** Unkomprimiert: $\mathrm{ld}(32)=5$ bit/Zeichen. Verhaeltnis $5/2{,}2$; Dauer $T = 20 \cdot (5/2{,}2) \approx 45{,}45\,\mathrm{s}$.
