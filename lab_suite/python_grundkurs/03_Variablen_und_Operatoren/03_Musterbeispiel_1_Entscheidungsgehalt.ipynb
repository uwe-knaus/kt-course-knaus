## Fragestellung

Was versteht man unter dem **Entscheidungsgehalt $G$** eines Signals? Wie groß ist der **gesamte Entscheidungsgehalt**, wenn **5 Symbole** aus einem **Signalvorrat von 512 Symbolen** ausgewählt werden?

## Antwort (aus dem Skriptum)

- **Entscheidungsgehalt $G$:** Anzahl der binären Entscheidungen bei der Auswahl eines Signals aus dem Signalvorrat $V$.

- Es gilt:
  $$G = \mathrm{ld}(V) = \log_2(V) \quad \text{[Bit]}$$

- **Einzelnes Symbol** aus $V = 512$:
  $$G = \mathrm{ld}(512) = 9 \,\mathrm{Bit}$$
  (denn $2^9 = 512$)

- **5 Symbole** (jeweils aus demselben Vorrat):
  $$G_\mathrm{gesamt} = 5 \cdot 9 = 45 \,\mathrm{Bit}$$

## Python: Rechengang umsetzen

Mit **Variablen** und **Operatoren** (inkl. `math.log2` für $\log_2$):

```python
import math

V = 512           # Signalvorrat
anzahl_symbole = 5

G_pro_symbol = math.log2(V)   # ld(V)
G_gesamt = anzahl_symbole * G_pro_symbol

print("V =", V)
print("G pro Symbol = ld(V) =", G_pro_symbol, "Bit")
print("5 Symbole => G_gesamt =", G_gesamt, "Bit")
```
