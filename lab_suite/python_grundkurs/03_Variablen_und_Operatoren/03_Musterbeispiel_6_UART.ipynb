## Fragestellung

Wie lange dauert die Übertragung eines Files mit **12860 Bytes**, wenn für die Übertragung ein **UART** mit folgender Einstellung verwendet wird:
- Baudrate **9600 bit/s**
- 1 Start-Bit, 1 Stop-Bit
- Kein Parity-Bit?

## Antwort (aus dem Skriptum)

- Pro Byte werden **10 Bits** übertragen (1 Start + 8 Daten + 1 Stop).

- Gesamt: $12860 \cdot 10 = 128600$ Bits.

- Übertragungszeit:
  $$T = \frac{\text{Bits}}{\text{Baudrate}} = \frac{128600\,\mathrm{bit}}{9600\,\mathrm{bit/s}} \approx 13{,}4\,\mathrm{s}$$
