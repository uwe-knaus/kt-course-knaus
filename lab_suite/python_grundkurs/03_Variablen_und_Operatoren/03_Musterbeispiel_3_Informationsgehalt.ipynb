## Fragestellung

Welchen **Informationsgehalt** hat ein Zeichen, das mit einer **Wahrscheinlichkeit von $1/10\,000$** auftritt?

## Antwort (aus dem Skriptum)

Der Informationsgehalt $h$ (in Bit) eines Zeichens mit Auftrittswahrscheinlichkeit $p$ ist:

$$h = \mathrm{ld}\left(\frac{1}{p}\right) = \mathrm{ld}(V) \quad \text{[Bit]}$$

mit $V = 1/p$ („effektiver Vorrat“).

Für $p = 1/10\,000$:

$$h = \mathrm{ld}(10\,000) = \frac{\log_{10}(10\,000)}{\log_{10}(2)} \approx 13{,}28 \,\mathrm{Bit}$$

*Ein seltenes Zeichen trägt mehr Information: Zur eindeutigen Auswahl sind etwa 13,28 binäre Entscheidungen nötig.*

## Python: Rechengang umsetzen

Variable $p$, Ausdruck $\log_2(1/p)$ mit `math.log2`:

```python
import math

p = 1 / 10_000   # Auftrittswahrscheinlichkeit des Zeichens

# Informationsgehalt h = ld(1/p) = log2(1/p)
h = math.log2(1 / p)

print("p =", p)
print("h = ld(1/p) =", h, "Bit")
```
