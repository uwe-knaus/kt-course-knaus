## Zuweisung: Name = Wert

Eine **Variable** ist ein Name, unter dem du einen Wert speicherst. Du weist mit **`=`** zu: `name = wert`. Danach kannst du den Namen verwenden, um auf den Wert zuzugreifen.

Der Typ (int, float, str, …) gehört zum **Wert**, nicht zur Variable – du deklarierst keinen Typ wie in C.

```python
a = 10
b = 20
print("a =", a, ", b =", b)
print("Summe:", a + b)
```

## Arithmetik: +, -, *, /, //, %, **

| Operator | Bedeutung | Beispiel |
|----------|------------|----------|
| `+` | Addition | `3 + 5` → 8 |
| `-` | Subtraktion | `10 - 4` → 6 |
| `*` | Multiplikation | `3 * 4` → 12 |
| `/` | Division (immer float) | `7 / 2` → 3.5 |
| `//` | Ganzzahldivision | `7 // 2` → 3 |
| `%` | Rest (Modulo) | `7 % 2` → 1 |
| `**` | Potenz | `2 ** 8` → 256 |

```python
print(7 / 2)
print(7 // 2)
print(7 % 2)
print("2^8 =", 2 ** 8)
```

## Vergleichsoperatoren: Ergebnis ist `bool`

Vergleiche liefern **`True`** oder **`False`** (Typ `bool`). Du brauchst sie später für Bedingungen (`if`) und Schleifen.

| Operator | Bedeutung |
|----------|------------|
| `==` | gleich |
| `!=` | ungleich |
| `<` | kleiner |
| `>` | größer |
| `<=` | kleiner oder gleich |
| `>=` | größer oder gleich |

```python
x = 5
print("x == 5:", x == 5)
print("x < 3:", x < 3)
print("x >= 5:", x >= 5)
```

## Ausdrücke: Variablen, Literale, Operatoren

Ein **Ausdruck** ist eine Kombination aus Variablen, Literalen (z. B. 42, 3.14) und Operatoren, die zu einem Wert ausgewertet wird. Du kannst Ausdrücke in Zuweisungen und in `print()` verwenden.

```python
M = 2
k = 8
signalvorrat = M ** k
print("Signalvorrat M^k =", signalvorrat)
```
