## Fragestellung

Wie groß ist der **Bezeichnungsraum R** für eine **5-stellige Dezimalzahl**? Wie groß wird dieser Bezeichnungsraum, wenn die **Basis B=2** (Binärdarstellung) verwendet wird?

## Antwort (aus dem Skriptum)

- **5-stellige Dezimalzahl** (0 … 99999):
  $$V = 100\,000$$

- **Bezeichnungsraum** (Basis B):
  $$R = B \cdot \log_{10}(V)$$

- **Dezimal (B=10):**
  $$R = 10 \cdot \log_{10}(100\,000) = 10 \cdot 5 = 50$$

- **Basis B=2:**
  $$R = 2 \cdot \mathrm{ld}(100\,000) = 2 \cdot \frac{\log_{10}(100\,000)}{\log_{10}(2)} \approx 33{,}2$$

## Python: Rechengang umsetzen

Variablen für V, B; math.log10 und math.log2 für die Logarithmen:

```python
import math

V = 100_000   # 5-stellige Dezimalzahl: 0 bis 99999

# Dezimal: R = B * log10(V), B=10
B_dez = 10
R_dez = B_dez * math.log10(V)
print("V =", V)
print("Dezimal (B=10): R =", B_dez, "* log10(V) =", R_dez)

# Basis 2: R = 2 * ld(V) = 2 * log2(V)
R_bin = 2 * math.log2(V)
print("Basis 2: R = 2 * ld(V) =", R_bin)
```
