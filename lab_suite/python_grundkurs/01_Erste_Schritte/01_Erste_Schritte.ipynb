## Erstes Programm

Das klassische erste Programm ist eine Zeile: etwas ausgeben. In Python geht das mit **`print(...)`**.

Führe die nächste Zelle aus (**Shift+Enter**). Das ist dein erstes „Programm“ in diesem Kurs.

```python
print("Hallo, KT-Kurs!")
```

Dasselbe kannst du auch als **Datei** speichern:

1. In einer Datei `hallo.py` den Inhalt `print("Hallo, KT-Kurs!")` speichern.
2. Im Terminal (in dem Ordner, in dem `hallo.py` liegt): `python hallo.py` ausführen.

Dann wird die Datei als **Skript** einmal von vorne bis hinten ausgeführt.

## Skript vs. interaktiv (REPL)

| Art | Beschreibung |
|-----|--------------|
| **Skript** | Code in einer `.py`-Datei. Du startest mit `python datei.py` – das gesamte Skript läuft durch und endet. |
| **Interaktiv / REPL** | Du gibst Code Zeile für Zeile ein (oder, im **Notebook**, Zelle für Zelle). Nach jeder Ausführung siehst du sofort das Ergebnis und kannst weiter tippen. „REPL“ = Read–Eval–Print Loop. |

**Notebooks** sind eine Mischform: Du schreibst Code in Zellen und führst sie nacheinander aus (interaktiv), kannst aber auch längere Abläufe in einer Zelle haben (wie ein kleines Skript). Ideal zum Lernen und Ausprobieren.

## Erste Syntax-Hinweise (für C-Kenner)

Wenn du schon **C** kennst, fallen dir gleich Unterschiede auf:

- **Kein Semikolon:** In Python beendest du eine Anweisung mit dem Zeilenende – kein `;` nötig.
- **Keine geschweiften Klammern für Blöcke:** Was in C zwischen `{ }` steht, wird in Python durch **Einrückung** (z. B. Leerzeichen oder Tabs) markiert. Ein Block ist alles, das gleich eingerückt ist.
- **Kein `main()`:** Ein Python-Skript startet einfach oben; es gibt keinen verpflichtenden Einstiegspunkt wie in C.

Mehr dazu – und weitere Gegenüberstellungen – im Notebook **[C und Python im Vergleich](01_Erste_Schritte_C_Python_Vergleich.ipynb)**.
