## 1. Anweisungen: Kein Semikolon

| C | Python |
|---|--------|
| Jede Anweisung endet mit `;` | **Kein Semikolon.** Das Zeilenende beendet die Anweisung. |

In Python ist ein Semikolon am Zeilenende optional (und unüblich); mehrere kurze Anweisungen in einer Zeile trennst du damit – brauchst du aber kaum.

````python
# Richtig in Python:
print("Hallo")
print(42)
````

## 2. Blöcke: Einrückung statt `{ }`

| C | Python |
|---|--------|
| Blöcke mit `{ }` | **Einrückung** definiert den Block. Alles, was gleich weit eingerückt ist, gehört zusammen. |

Du musst konsistent sein: entweder überall Leerzeichen (z. B. 4) oder überall Tabs. Gemischt führt zu Fehlern.

````python
# Beispiel: if-Block (if kommt erst in Kapitel 05)
# In C: if (x > 0) { printf("positiv"); }
# In Python:
x = 5
if x > 0:
    print("positiv")   # diese Zeile ist der "Block" – eingerückt
````

## 3. Einstiegspunkt: Kein `main()`

| C | Python |
|---|--------|
| Programm startet in `main()` | Das Skript startet **oben**. Die erste Zeile wird zuerst ausgeführt. |

Wenn du trotzdem eine Art `main` willst (z. B. für bessere Lesbarkeit), kannst du später `if __name__ == "__main__":` nutzen – dazu in einem späteren Kapitel.

## 4. Variablen: Kein Typ bei der Deklaration

| C | Python |
|---|--------|
| `int x;` oder `double y = 3.14;` | **Kein Typ beim Schreiben.** Du weist zu: `x = 42` oder `y = 3.14`. Der Typ gehört zum Wert, nicht zur „Deklaration“. |

Variablen lernst du in **03_Variablen_und_Operatoren** – hier nur der Hinweis: Kein `int`, `float`, `char` vor dem Namen.

````python
# In C: int a = 10;
# In Python:
a = 10
print(a)
````

## 5. Ausgabe: `print()` statt `printf`

| C | Python |
|---|--------|
| `printf("Zahl: %d", x);` mit Formatzeichen | `print("Zahl:", x)` – mehrere Argumente durch Komma, Leerzeichen dazwischen. Oder f-Strings: `print(f"Zahl: {x}")` (später). |

````python
x = 42
print("Zahl:", x)
print("Zahl:", x, "und fertig")
````

## 6. Kommentare

| C | Python |
|---|--------|
| `//` einzeilig, `/* ... */` mehrzeilig | **`#`** für den Rest der Zeile. Mehrzeilige Kommentare oft als mehrzeiliger String `"""..."""` (wird ignoriert, wenn nicht zugewiesen). |

````python
# Das ist ein Kommentar in Python
print("sieht du")  # Kommentar am Zeilenende
````

## 7. Ein kurzes Programm: dieselbe Logik in C und Python

Zum Überblick: dasselbe Programm – zwei Zahlen addieren und ausgeben – zuerst in C, darunter in Python. So siehst du die Unterschiede (Semikolon, `main`, Typen, `printf` vs. `print`) in einem Stück.

**C:**
```c
#include <stdio.h>
int main(void) {
    int a = 10;
    int b = 20;
    int summe = a + b;
    printf("Summe: %d\n", summe);
    return 0;
}
```

**Python (gleiche Logik):**

````python
a = 10
b = 20
summe = a + b
print("Summe:", summe)
````
