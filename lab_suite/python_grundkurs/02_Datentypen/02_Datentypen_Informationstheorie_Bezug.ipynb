## Warum dieses Thema hier?

In der Vorlesung **2. Informationstheorie** geht es unter anderem um:

- **Signalvorrat:** Wie viele verschiedene Nachrichten sind mit M Zeichen und k Stellen möglich? (Formel: M^k)
- **Entropie:** Maß für den mittleren Informationsgehalt (in Bit).
- **Mittlerer Binärstellenaufwand:** Wie viele Binärstellen braucht man im Mittel pro Zeichen?

All das sind **Zahlenrechnungen**. Genau dafür brauchst du in Python die Datentypen **`int`** und **`float`**: ganze Zahlen für Anzahlen (z. B. Signalvorrat), Dezimalzahlen für Entropie und Mittelwerte. So verbindet sich das Python-Kapitel **Datentypen** direkt mit dem ersten KT-Themenblock.
