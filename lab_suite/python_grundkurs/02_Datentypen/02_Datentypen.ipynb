## Zahlen: `int` und `float`

- **`int`** = ganze Zahlen (z. B. 0, 42, -7).
- **`float`** = Dezimalzahlen (z. B. 3.14, -0.5, 2.0).

Du schreibst sie einfach hin; Python erkennt den Typ automatisch. Führe die nächste Zelle aus.

```python
print(42)
print(3.14)
print(type(42))    # int
print(type(3.14))  # float
```

## Strings: `str`

**Strings** sind Texte. Du gibst sie in Anführungszeichen ein – einfache `'...'` oder doppelte `"..."`.

```python
print("Hallo")
print('Auch so möglich')
print(type("Hallo"))  # str
```

## Boolesche Werte: `True` und `False`

Für Bedingungen und Vergleiche brauchst du **Wahrheitswerte**: `True` und `False` (mit großem Anfangsbuchstaben). Sie kommen z. B. bei „ist 5 größer als 3?“ vor – dazu später in **03_Variablen_und_Operatoren**.

```python
print(True)
print(False)
print(type(True))  # bool
```

## Typen prüfen mit `type(...)`

Mit **`type(Wert)`** siehst du, welcher Datentyp vorliegt. Das hilft beim Experimentieren und beim Debuggen.

```python
print(type(0))
print(type(1.0))
print(type("0"))   # Achtung: "0" ist ein String, keine Zahl!
```
