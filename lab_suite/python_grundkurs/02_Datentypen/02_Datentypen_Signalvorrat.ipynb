## Formel: Signalvorrat = M^k

In der Vorlesung: **Signalvorrat N** = Anzahl der möglichen verschiedenen Nachrichten, wenn du **M** Zeichen (Alphabet) hast und jede Nachricht aus **k** Zeichen besteht:

N = M^k

Beispiele:
- Binär (M=2), 8 Stellen (k=8): N = 2^8 = 256.
- Dezimalziffern (M=10), 3 Stellen: N = 10^3 = 1000.

In Python rechnest du das mit **ganzen Zahlen** (int) und dem Potenz-Operator **`**`**.

```python
# Signalvorrat N = M^k
M = 2   # Binär
k = 8
N = M ** k
print("Binär, 8 Stellen: N =", N)
```

(Hinweis: Variablen wie M, k, N lernst du offiziell in **03_Variablen_und_Operatoren**. Hier nutzen wir sie schon ganz knapp, damit die Formel lesbar wird.)

```python
# Dezimalziffern (0–9), 3 Stellen
print("Dezimal, 3 Stellen: N =", 10 ** 3)

# Noch ein Beispiel: 26 Buchstaben, Wortlänge 2
print("26 Zeichen, 2 Stellen: N =", 26 ** 2)
```
