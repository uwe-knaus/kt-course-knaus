## Python vs. C++: Übersicht

| Konzept | C++ | Python |
|---------|-----|--------|
| **Klasse** | `class Name { ... };` | `class Name:` mit eingerücktem Rumpf |
| **Konstruktor** | `Name(...) { ... }` | `def __init__(self, ...):` |
| **this** | `this->member` bzw. `this->methode()` | **`self`** – wird explizit als erster Parameter übergeben |
| **Methoden** | Funktionsdeklaration in der Klasse | `def methode(self, ...):` (Einrückung unter der Klasse) |
| **Vererbung** | `class Kind : public Eltern { ... };` | `class Kind(Eltern):` |
| **Instanz erzeugen** | `Name obj(args);` oder `Name* p = new Name(args);` | `obj = Name(args)` – kein `new` |

In Python gibt es **kein** `new`; du „rufst" die Klasse wie eine Funktion auf, und Python erzeugt das Objekt und ruft `__init__` auf.

## Klasse definieren und Objekt instanziieren

Eine **Klasse** ist der Bauplan; eine **Instanz** (Objekt) ist ein konkretes Exemplar. Du definierst eine Klasse mit **`class Name:`**; der **Konstruktor** heißt **`__init__(self, ...)`**. Das **`self`** bezeichnet die Instanz selbst (wie `this` in C++).

````python
class Messwert:
    def __init__(self, wert, einheit="V"):
        self.wert = wert
        self.einheit = einheit

    def als_string(self):
        return f"{self.wert} {self.einheit}"

# Objekt instanziieren (kein 'new')
m = Messwert(3.14, "V")
print(m.als_string())
print(m.wert, m.einheit)
````

**C++-Entsprechung (Konzept):**

```cpp
class Messwert {
public:
    double wert;
    std::string einheit;
    Messwert(double w, const std::string& e = "V") : wert(w), einheit(e) {}
    std::string als_string() const { return std::to_string(wert) + " " + einheit; }
};
Messwert m(3.14, "V");
```

In Python werden Attribute typischerweise im `__init__` gesetzt; es gibt keine getrennte Deklaration wie in C++.

## Methoden

**Methoden** sind Funktionen, die zur Klasse gehören. Die **erste Parameter** ist immer **`self`** (die Instanz). Beim Aufruf übergibst du `self` nicht – Python setzt das Objekt automatisch ein.

| Python | C++ (vereinfacht) |
|--------|------------------|
| `def f(self, x):` | `void f(T x);` – `this` ist implizit |
| `obj.f(5)` | `obj.f(5);` |

Ohne `self` wäre es eine **Klassenmethode** oder **statische Methode** (entspricht grob `static` in C++); für den Einstieg reicht die normale Instanzmethode.

````python
class Punkt:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def abstand_ursprung(self):
        return (self.x**2 + self.y**2) ** 0.5

    def __str__(self):
        return f"P({self.x}, {self.y})"

p = Punkt(3, 4)
print(p)
print("Abstand:", p.abstand_ursprung())
````

`__str__` ist eine **spezielle Methode**: Sie wird aufgerufen, wenn du das Objekt mit `print()` oder in einem String ausgibst (vergleichbar mit Operator-Überladung / Ausgabe in C++).

## Vererbung (Klassenvererbung)

Eine **Kindklasse** erbt von einer **Basisklasse**: Sie übernimmt Attribute und Methoden und kann sie erweitern oder überschreiben. In Python schreibst du die Elternklasse in Klammern: **`class Kind(Eltern):`**.

**C++:** `class Kind : public Eltern { ... };`  
**Python:** `class Kind(Eltern):`

Um die Methode der **Elternklasse** aufzurufen, nutzt du **`super()`** (entspricht grob dem Aufruf der Basisklassen-Version in C++).

````python
class Signal:
    def __init__(self, name):
        self.name = name

    def beschreibung(self):
        return f"Signal: {self.name}"

class SinusSignal(Signal):
    def __init__(self, name, frequenz):
        super().__init__(name)
        self.frequenz = frequenz

    def beschreibung(self):
        return super().beschreibung() + f", f = {self.frequenz} Hz"

s = SinusSignal("Ton", 440)
print(s.beschreibung())
print(s.name, s.frequenz)
````

**Wichtig:** In `SinusSignal.__init__` rufst du **`super().__init__(name)`** auf, damit die Basisklasse ihre Initialisierung ausführt (wie im C++-Konstruktor der Aufruf des Eltern-Konstruktors). Ohne diesen Aufruf wäre `self.name` nicht gesetzt.

## Kurz zusammengefasst

| Thema | Inhalt |
|-------|--------|
| **Klasse** | `class Name:` mit eingerücktem Rumpf; Konstruktor: `__init__(self, ...)`. |
| **Instanziierung** | `obj = Name(args)` – kein `new`; Python ruft `__init__` auf. |
| **self** | Erster Parameter von Methoden; bezeichnet die Instanz (wie `this` in C++). |
| **Methoden** | `def methode(self, ...):`; Aufruf: `obj.methode(...)`. |
| **Vererbung** | `class Kind(Eltern):`; Eltern-Konstruktor: `super().__init__(...)`; Methode überschreiben und ggf. `super().methode()` aufrufen. |
