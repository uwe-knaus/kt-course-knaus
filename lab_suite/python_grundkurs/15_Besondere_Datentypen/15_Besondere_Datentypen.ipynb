## Der Typ `None`

**`None`** ist der einzige Wert vom Typ **`NoneType`**. Er bedeutet: **„kein Wert"** oder **„nichts gesetzt"**. Du nutzt ihn z. B.:

- als **Rückgabewert** einer Funktion, die „nichts" zurückgibt (entspricht grob `void` in C);
- als **Platzhalter** für „noch unbekannt" oder „optional nicht vorhanden";
- zum **Vergleichen**: `if x is None:` (nicht `== None` – üblich ist `is None`).

```python
x = None
print(type(x))
print(x is None)

def suche(wert, liste):
    if wert in liste:
        return liste.index(wert)
    return None

print(suche(3, [1, 2, 3]))
print(suche(5, [1, 2, 3]))
```

## Bytes: `bytes` und `bytearray`

**`bytes`** ist eine **unveränderliche** Folge von Bytes (0–255). **`bytearray`** ist die **veränderliche** Variante. Du brauchst sie für:

- **Binärdateien** (siehe 14_Dateien: Modi `"rb"`/`"wb"`);
- **Übertragung/Codierung**: Zeichen werden als Bytefolge gesendet (z. B. UTF-8, ASCII).

Erzeugen: `bytes([1, 2, 3])`, `b"Hello"` (Literal für bytes), oder aus String mit **`encode()`**.

```python
b = bytes([72, 101, 108, 108, 111])
print(b)
print(b"Hallo")

ba = bytearray(b)
ba[0] = 104
print(ba)
```

## Umwandlung: `encode()` und `decode()`

**Zeichenkette → Bytes:** **`str.encode(encoding)`** – z. B. `"Hallo".encode("utf-8")` oder `"Hallo".encode("ascii")`. Standard ist oft UTF-8.

**Bytes → Zeichenkette:** **`bytes.decode(encoding)`** – z. B. `b"Hallo".decode("utf-8")`. Nur gültige Bytefolgen für die gewählte Kodierung liefern sinnvollen Text.

So kannst du Text für Dateien oder Übertragung in Bytes umwandeln und beim Lesen wieder in Strings decodieren – wichtig für Codierung (KT-Kapitel 01, 08).

```python
text = "Hallo KT"
gepackt = text.encode("utf-8")
print("Bytes:", gepackt)

wieder_text = gepackt.decode("utf-8")
print("Zurück:", wieder_text)

print("ASCII:", "AB".encode("ascii"))
```

## Kurz: `set` (Menge)

Ein **`set`** ist eine **ungeordnete** Sammlung **eindeutiger** Elemente (keine Duplikate). Nützlich z. B. für „alle vorkommenden Symbole" oder „eindeutige Werte".

Erzeugen: `{1, 2, 3}` oder `set([1, 2, 2, 3])` → `{1, 2, 3}`.

```python
symbole = set("abacaba")
print("Eindeutige Zeichen:", symbole)
print("Anzahl verschiedener Zeichen:", len(symbole))
```

## Kurz zusammengefasst

| Typ | Bedeutung / Anwendung |
|-----|------------------------|
| **None** | „Kein Wert"; optionale Rückgabe, Platzhalter; Vergleichen mit `is None`. |
| **bytes** | Unveränderliche Bytefolge; Literal `b"..."`; Binärdaten, encode. |
| **bytearray** | Veränderliche Bytefolge. |
| **encode()** | String → Bytes (z. B. `"text".encode("utf-8")`). |
| **decode()** | Bytes → String (z. B. `daten.decode("utf-8")`). |
| **set** | Menge eindeutiger Elemente, ungeordnet. |

Damit kannst du mit Binärdaten und Codierung arbeiten – passend zu Dateien (14) und zu Codierungsthemen in der KT-Vorlesung.
