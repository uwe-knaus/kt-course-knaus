# 10 – Tupel

Ein **Tupel** (engl. *tuple*) ist eine **geordnete, unveränderliche** Folge von Werten. Du brauchst es z. B. für **mehrere Rückgabewerte** von Funktionen (Kapitel 11). Hier lernst du das Konzept kennen, den **Unterschied zur Liste**, ob es in C etwas Vergleichbares gibt, und wo Tupel in Python sinnvoll eingesetzt werden.

## Was ist ein Tupel?

Ein Tupel wird mit **runden Klammern** und Kommas geschrieben: `(3, 5)` oder `("a", "b", 42)`. Die Elemente haben eine **Reihenfolge** und sind über einen **Index** zugreifbar – ähnlich wie bei einer Liste. Der entscheidende Unterschied: Ein Tupel ist **unveränderlich** (immutable). Du kannst nach der Erstellung **kein** Element ändern, hinzufügen oder entfernen.

```python
paar = (10, 20)
print(paar)
print(paar[0], paar[1])

# paar[0] = 99   # Fehler! Tupel sind unveränderlich.
```

## Tupel vs. Liste

| | **Liste** | **Tupel** |
|---|-----------|----------|
| Schreibweise | `[a, b, c]` (eckige Klammern) | `(a, b, c)` (runde Klammern) |
| Verändern | Ja: `liste[0] = 5`, `liste.append(7)` | **Nein** – nach Erstellung fest. |
| Typ | `list` | `tuple` |
| Typischer Einsatz | Sammlung, die wächst oder sich ändert (z. B. Messwerte) | Feste Gruppe von Werten (z. B. Koordinaten, Rückgabewerte). |

Wenn du eine **feste Anzahl** von Werten zusammenhalten willst und sie sich **nicht ändern** sollen, ist ein Tupel oft passender und macht die Absicht klar.

```python
liste = [1, 2, 3]
liste[0] = 99
liste.append(4)
print(liste)

tupel = (1, 2, 3)
print(tupel)
# tupel[0] = 99  # würde Fehler auslösen
```

## Gibt es Tupel in C?

**Nein.** C hat keinen eingebauten Typ „Tupel". Du kommst dem Konzept am nächsten mit:

- **struct:** Eine feste Gruppe von Feldern (z. B. `struct Punkt { double x; double y; };`). Die „Reihenfolge" und „feste Anzahl" sind ähnlich; ein struct ist aber ein eigener Typ mit benannten Feldern.
- **Array mit fester Länge:** z. B. `int arr[2];` – aber ohne die Semantik „diese Werte gehören untrennbar zusammen" und ohne einfaches „Entpacken" wie in Python.

In Python ist das Tupel **eingebaut** und wird z. B. genutzt, um **mehrere Rückgabewerte** einer Funktion in einem Schritt zurückzugeben – in C würdest du dafür Zeiger-Parameter oder ein struct verwenden.

## Wo sind Tupel sinnvoll?

1. **Mehrere Rückgabewerte:** `return a, b` liefert ein Tupel `(a, b)`; beim Aufruf: `x, y = funktion()`. Das kommt in **11_Funktionen** ausführlich vor.
2. **Feste Paare / Koordinaten:** z. B. `(x, y)` für einen Punkt – die beiden Werte gehören zusammen und ändern sich nicht als „Einheit".
3. **Schlüssel in Wörterbüchern:** Dict-Keys müssen unveränderlich sein; Tupel aus unveränderlichen Werten sind erlaubt (z. B. `{(1, 2): "Wert"}`).
4. **Austausch von „Paketen":** Wenn du genau zwei oder drei Werte gemeinsam übergeben oder zurückgeben willst, ohne eine Liste zu verändern, ist ein Tupel passend.

```python
# Beispiel: Koordinaten als Tupel
punkt = (2.5, 1.0)
x, y = punkt
print("x =", x, "y =", y)

# Entpacken beim Zuweisen
a, b = (10, 20)
print("a =", a, "b =", b)
```

```python
# Beispiel: „Paket" von Werten (z. B. Min, Max, Mittel)
def min_max_mittel(werte):
    return min(werte), max(werte), sum(werte) / len(werte)

zahlen = [3, 1, 4, 1, 5]
mini, maxi, mittel = min_max_mittel(zahlen)
print("Min:", mini, "Max:", maxi, "Mittel:", mittel)
```

## Kurz zusammengefasst

| Thema | Inhalt |
|-------|--------|
| **Tupel** | Geordnete, **unveränderliche** Folge: `(a, b, c)`; Zugriff per Index. |
| **vs. Liste** | Liste `[]` ist veränderlich; Tupel `()` ist fest. |
| **C** | Kein Tupel-Typ; Annäherung: struct oder festes Array. |
| **Anwendung** | Mehrere Rückgabewerte, Koordinaten/Paare, Dict-Keys, feste Wertpakete. |

Mit diesem Konzept verstehst du in **11_Funktionen**, warum `return a, b` und `x, y = f()` funktionieren – es ist ein Tupel, das beim Zuweisen entpackt wird.
