# 04 – Listen und Sequenzen

Hier lernst du **Listen** (`list`) kennen: mehrere Werte in einer geordneten Reihenfolge speichern, per **Index** oder **Slice** darauf zugreifen und mit einer **for-Schleife** darüber laufen. Listen sind die Grundlage für Folgen von Daten – z. B. Messwerte, Symbolfolgen oder später Arrays in der Signalverarbeitung.

## Liste erstellen: `[]` und Kommas

Eine **Liste** ist eine geordnete Sammlung von Elementen. Du erzeugst sie mit **eckigen Klammern** `[]`; die Elemente trennst du mit **Kommas**. Die Elemente können Zahlen, Strings oder auch andere Typen sein.

```python
zahlen = [3, 7, 2, 9, 1]
namen = ["Alice", "Bob", "Charlie"]
gemischt = [10, "Text", 3.14, True]

print("Zahlen:", zahlen)
print("Namen:", namen)
print("Gemischt:", gemischt)
```

## Index: ein einzelnes Element holen

Du greifst auf ein Element zu, indem du seinen **Index** in eckige Klammern schreibst. Der **erste** Index ist **0** (nicht 1 wie in der Umgangssprache).

**Negative Indizes** zählen vom Ende: `-1` = letztes Element, `-2` = vorletztes usw.

```python
zahlen = [3, 7, 2, 9, 1]

print("zahlen[0] =", zahlen[0])   # erstes Element
print("zahlen[2] =", zahlen[2])   # drittes Element
print("zahlen[-1] =", zahlen[-1]) # letztes Element
print("zahlen[-2] =", zahlen[-2]) # vorletztes
```

## Slice: einen Bereich auswählen

Mit einem **Slice** `[start:end]` holst du einen **Bereich** aus der Liste. Es werden die Elemente von Index `start` **bis vor** `end` genommen (also `end` ist nicht mehr dabei).

- `liste[start:end]` – von `start` bis vor `end`
- `liste[:end]` – vom Anfang bis vor `end`
- `liste[start:]` – von `start` bis Ende
- `liste[start:end:step]` – mit Schrittweite `step` (z. B. jedes 2. Element)

```python
zahlen = [3, 7, 2, 9, 1, 4, 6]

print("zahlen[1:4] =", zahlen[1:4])   # Indizes 1, 2, 3
print("zahlen[:3] =", zahlen[:3])     # erste drei
print("zahlen[4:] =", zahlen[4:])     # ab Index 4 bis Ende
print("zahlen[::2] =", zahlen[::2])   # jedes 2. Element
```

## Länge: `len(liste)`

Die Funktion **`len(...)`** liefert die **Anzahl der Elemente** in einer Liste (oder in einem String).

```python
zahlen = [3, 7, 2, 9, 1]
print("Anzahl Elemente:", len(zahlen))
print("Letztes Element (über len):", zahlen[len(zahlen) - 1])
print("Oder einfacher mit -1:", zahlen[-1])
```

## for-Schleife: über alle Elemente iterieren

Mit **`for element in liste:`** gehst du **der Reihe nach** durch alle Elemente. Der Code unter der Zeile mit `for` (der eingerückte Block) wird für jedes Element einmal ausgeführt.

```python
namen = ["Alice", "Bob", "Charlie"]

for name in namen:
    print("Hallo,", name)
```

```python
zahlen = [3, 7, 2, 9, 1]
summe = 0
for z in zahlen:
    summe = summe + z
print("Summe:", summe)
```

## Kurz zusammengefasst

| Konzept | Syntax / Beispiel |
|--------|-------------------|
| Liste erstellen | `[1, 2, 3]`, `["a", "b"]` |
| Ein Element | `liste[0]`, `liste[-1]` |
| Bereich (Slice) | `liste[1:4]`, `liste[:3]`, `liste[::2]` |
| Länge | `len(liste)` |
| Schleife | `for x in liste:` … |

Listen sind die Basis für **Folgen von Werten** – z. B. Zeichenketten (Strings sind ebenfalls Sequenzen), Messreihen oder später Arrays in der Signal- und Spektrenanalyse.
