## Funktion definieren: Python vs. C

| C | Python |
|---|--------|
| Rückgabetyp und Name: `double f(double x) { ... }` | **`def`** Name(Parameter): Rumpf eingerückt; **kein** Typ in der Signatur. |
| Block mit `{ }` | Block durch **Einrückung** (Doppelpunkt nach der Parameterliste). |
| `return wert;` (ein Wert, feste Typen) | `return wert` – ein Wert; **oder** mehrere Werte als Tupel (siehe unten). |

In Python schreibst du **`def funktionsname(parameter):`** und darunter den eingerückten Code. Es gibt **keine** Deklaration von Rückgabetyp oder Parametertypen (optional mit Type Hints, aber nicht Pflicht).

````python
def quadrat(x):
    return x * x

print(quadrat(5))   # 25
````

## Rückgabewert: `return`

Mit **`return ausdruck`** beendest du die Funktion und gibst einen **Wert** zurück. Ohne `return` oder mit `return` ohne Argument liefert die Funktion **`None`** (entspricht grob „nichts" / void in C).

**C:** Es gibt genau **einen** Rückgabetyp; `void` für „nichts". **Python:** Es gibt immer „einen" Rückgabewert – der kann aber ein **Tupel** sein, also mehrere Werte in einem.

````python
def min_max(a, b):
    if a <= b:
        return a, b   # zwei Werte als Tupel
    return b, a

kleiner, groesser = min_max(10, 3)   # Entpacken
print("Kleiner:", kleiner, "Größer:", groesser)
````

## Mehrere Rückgabewerte

Eine Python-Funktion kann **mehrere Werte** zurückgeben: Du schreibst z. B. **`return a, b`**. Intern ist das ein **Tupel** `(a, b)`. Beim Aufruf kannst du die Werte **entpacken**: `x, y = funktion()`.

**C:** Dafür brauchst du Zeiger/Referenzen (Ausgabeparameter) oder ein Struct. In Python ist das mit einem einfachen `return a, b` und `x, y = f()` erledigt.

## Dieselbe Logik in C und Python: Funktion für Sinus + optional AWGN

**Aufgabe (wie in 09_Steuerung):** Wir erzeugen Sinussignal und AWGN. Statt die Verzweigung im Hauptprogramm zu machen, rufen wir eine **Funktion** auf: Sie bekommt **sinus**, **awgn** und **NOISY**; sie gibt das **entsprechende Signal** zurück (nur Sinus oder Sinus + Rauschen). Danach plotten wir das zurückgegebene Signal.

### C-Variante (Konzept)

In C hätte man z. B. eine Funktion, die das Ergebnis in ein übergebenes Array schreibt oder einen Zeiger auf das „aktuelle" Array zurückgibt:

```c
// Signatur z.B.: void get_signal(double *sinus, double *awgn, int noisy, double *out);
// oder: double* get_signal(...);  mit festem Puffer

void get_signal(double *sinus, double *awgn, int noisy, double *out, int n) {
    if (noisy)
        for (int i = 0; i < n; i++) out[i] = sinus[i] + awgn[i];
    else
        for (int i = 0; i < n; i++) out[i] = sinus[i];
}

// Aufruf: get_signal(sinus, awgn, NOISY, signal, 100);
// dann plot(signal);
```

C kann nur **einen** Wert zurückgeben (oder void); die „Rückgabe" eines Arrays erfolgt über Zeiger/out-Parameter. In Python geben wir direkt ein Array (NumPy-Array) zurück.

### Python: Funktion liefert Signal, Hauptprogramm plottet

Die Funktion **`get_signal(sinus, awgn, noisy)`** gibt abhängig von **noisy** entweder nur **sinus** oder **sinus + awgn** zurück. Wir generieren danach wie in Kapitel 09 Sinus und AWGN, rufen die Funktion auf und plotten das Ergebnis.

````python
import numpy as np
import matplotlib.pyplot as plt

def get_signal(sinus, awgn, noisy):
    """Liefert Sinus oder Sinus+AWGN je nach noisy (bool)."""
    if noisy:
        return sinus + awgn
    return sinus

NOISY = True   # True = verrauschtes Signal, False = nur Sinus

n = 100
t = np.linspace(0, 1, n)
f = 3
sinus = np.sin(2 * np.pi * f * t)
np.random.seed(42)
awgn = 0.3 * np.random.randn(n)

signal = get_signal(sinus, awgn, NOISY)

plt.figure(figsize=(8, 3))
if NOISY:
    plt.plot(t, signal, color="green", label="Sinus + AWGN")
    plt.title("Verrauschtes Signal (NOISY = True)")
else:
    plt.plot(t, signal, color="blue", label="Sinus")
    plt.title("Reines Sinussignal (NOISY = False)")
plt.xlabel("t")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.show()
````

Du kannst **NOISY** umstellen und die Zelle erneut ausführen – die Funktion liefert jedes Mal das passende Signal, das wir anschließend plotten.

## Kurz zusammengefasst

| Thema | Inhalt |
|-------|--------|
| **Definition** | `def name(parameter):` mit eingerücktem Rumpf; kein Rückgabetyp in der Signatur. |
| **return** | Ein Wert mit `return wert`; ohne return liefert die Funktion `None`. |
| **Mehrere Rückgaben** | `return a, b` (Tupel); Aufruf mit `x, y = f()`. In C: Zeiger oder Struct. |
| **Beispiel** | `get_signal(sinus, awgn, NOISY)` gibt Sinus oder Sinus+AWGN zurück; Hauptprogramm plottet. |

Damit hast du Funktionen mit klarer Verantwortung: Die Funktion entscheidet nur, welches Signal zurückgegeben wird; das Plotten bleibt im aufrufenden Code.
