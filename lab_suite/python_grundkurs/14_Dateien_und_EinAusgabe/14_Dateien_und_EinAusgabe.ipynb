## C vs. Python: Dateizugriff im Überblick

| Konzept | C | Python |
|---------|---|--------|
| **Handle/Objekt** | `FILE* fp;` | `f = open(...)` liefert ein Dateiobjekt |
| **Öffnen** | `fopen("datei.txt", "r")` / `"w"` / `"rb"` | `open("datei.txt", "r")` / `"w"` / `"rb"` |
| **Lesen (Text)** | `fgets(buf, n, fp)`, `fscanf` | `f.read()`, `f.readline()`, `f.readlines()` |
| **Schreiben (Text)** | `fprintf(fp, ...)`, `fputs` | `f.write(s)` |
| **Binär** | `fread` / `fwrite` mit `"rb"` / `"wb"` | `open(..., "rb")` / `"wb"`; `f.read(n)` Bytes, `f.write(bytes)` |
| **Schließen** | `fclose(fp)` – wichtig, sonst Ressourcenleck | `f.close()` oder besser **Context Manager** (siehe unten) |

Die Modi (`"r"`, `"w"`, `"a"`, `"rb"`, `"wb"`) sind in beiden Sprachen ähnlich; in Python wird das Dateiobjekt nach Gebrauch geschlossen, idealerweise mit **`with`**.

## Textdateien (ASCII/UTF-8) und Binärdateien

- **Textdateien:** Inhalt ist **Zeichen** (z. B. ASCII oder UTF-8). Du liest/schreibst **Strings**. Modi: **`"r"`** (lesen), **`"w"`** (schreiben, überschreibt), **`"a"`** (anhängen). Python kümmert sich um Kodierung (Standard: UTF-8).
- **Binärdateien:** Inhalt sind **rohe Bytes** (z. B. Bilder, Messdaten als Zahlen). Modi: **`"rb"`** (lesen), **`"wb"`** (schreiben). Du arbeitest mit **`bytes`** bzw. `bytearray`; keine automatische Zeichenumsetzung.

**C:** Dort gibt es ebenfalls Text- (`fgets`, `fprintf`) vs. Binärmodus (`fread`/`fwrite` mit `"rb"`/`"wb"` bei `fopen`). In Python ist die Trennung durch die Modi und den Datentyp (str vs. bytes) klar.

## Datei öffnen, lesen, schreiben, schließen

**Grundablauf:** `open(Pfad, Modus)` → lesen/schreiben → `close()`. Ohne `close()` bleibt die Datei geöffnet (Ressourcen, Puffer werden unter Umständen nicht geschrieben).

```python
# Textdatei schreiben (wie in C: fopen mit "w", dann write/fprintf, dann fclose)
f = open("beispiel.txt", "w")
f.write("Erste Zeile\n")
f.write("Zweite Zeile\n")
f.close()

# Textdatei lesen
f = open("beispiel.txt", "r")
inhalt = f.read()
f.close()
print(inhalt)
```

```python
# Zeile für Zeile: readline() bzw. readlines()
f = open("beispiel.txt", "r")
zeile1 = f.readline()
zeile2 = f.readline()
f.close()
print("Zeile 1:", zeile1.strip())
print("Zeile 2:", zeile2.strip())
```

## Context Manager: `with open(...) as f:`

Wenn du **`with open(...) as f:`** verwendest, wird die Datei am **Ende des Blocks** (oder bei einer Exception) **automatisch geschlossen**. Du musst **`close()`** nicht mehr selbst aufrufen – und vergisst es auch nicht.

**C:** Es gibt keine eingebaute Entsprechung; man muss immer `fclose` aufrufen (oder z. B. in C++ RAII mit eigenen Wrapper-Klassen nutzen). In Python ist der Context Manager die empfohlene Art, mit Dateien zu arbeiten.

```python
with open("beispiel.txt", "r") as f:
    inhalt = f.read()
# Datei ist hier bereits geschlossen
print(inhalt)
```

```python
# Schreiben mit with – Puffer wird beim Verlassen des Blocks geleert, Datei geschlossen
with open("ausgabe.txt", "w") as f:
    f.write("Hallo\n")
    f.write("Welt\n")
# f ist geschlossen
print("Datei ausgabe.txt geschrieben.")
```

## Binärdateien: Bytes lesen und schreiben

Mit **`"rb"`** und **`"wb"`** arbeitest du mit **Bytes**. Nützlich für Dateiformate, die keine Textzeilen sind (z. B. Rohdaten, Bilder).

```python
# Einige Bytes schreiben
daten = bytes([0x48, 0x65, 0x6C, 0x6C, 0x6F])  # "Hello" in ASCII
with open("binary.bin", "wb") as f:
    f.write(daten)

# Binär lesen
with open("binary.bin", "rb") as f:
    gelesen = f.read()
print("Gelesene Bytes:", gelesen)
print("Als Dekodierung (ASCII):", gelesen.decode("ascii"))
```

## Konsole: `input()` und `print()`

**Eingabe von der Konsole:** **`input("Prompt: ")`** – liefert eine Zeile als String (wie in C z. B. `fgets` von stdin oder `scanf`).  
**Ausgabe:** **`print(...)`** kennst du bereits (vergleichbar mit `printf`).

```python
# input() – in Jupyter/Notebook: Eingabefeld erscheint
name = input("Dein Name: ")
print("Hallo,", name)
```

## Kurz zusammengefasst

| Thema | Inhalt |
|-------|--------|
| **C-Vergleich** | `FILE*`, `fopen`/`fclose`; Text mit `fgets`/`fprintf`, Binär mit `fread`/`fwrite`. Python: `open()`, Modi wie in C. |
| **Textdateien** | Modi `"r"`, `"w"`, `"a"`; `read()`, `readline()`, `readlines()`, `write(s)`. |
| **Binärdateien** | Modi `"rb"`, `"wb"`; Lesen/Schreiben von `bytes`. |
| **Context Manager** | `with open(...) as f:` – Datei wird am Blockende automatisch geschlossen; empfohlene Praxis. |
| **Konsole** | `input()` für Eingabe, `print()` für Ausgabe. |

Damit kannst du in Python Dateien sicher öffnen und schließen und sowohl Text- als auch Binärdaten lesen und schreiben.
