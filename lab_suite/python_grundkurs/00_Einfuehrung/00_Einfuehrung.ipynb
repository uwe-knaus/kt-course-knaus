## Was ist Python?

Python ist eine **interpretierte**, **vielseitige** Programmiersprache:

- **Interpretiert:** Das Programm wird zur Laufzeit übersetzt und ausgeführt – du musst nicht extra kompilieren.
- **Vielseitig:** Python wird in Wissenschaft, Datenanalyse, Web, Automatisierung und vielen anderen Bereichen eingesetzt.

In der Kommunikationstechnik nutzen wir Python z. B. für Berechnungen (Entropie, Signale), für Visualisierungen (Plots, Spektren) und für die Labor-Software (z. B. die KT-Lab-Suite).

## Wann wurde Python entwickelt?

- **Anfang der 1990er Jahre:** Guido van Rossum entwickelte Python am CWI in Amsterdam.
- **1991:** Erste Veröffentlichung.
- **Python 2 (2000) und Python 3 (2008):** Größere Sprachänderungen; heute wird **Python 3** verwendet. Python 2 ist veraltet.

## Warum heißt es Python?

Der Name bezieht sich **nicht** auf die Schlange, sondern auf die britische Comedy-Truppe **Monty Python**. Guido van Rossum war ein Fan und wählte einen kurzen, einprägsamen Namen. Die Dokumentation und viele Beispiele im Python-Umfeld enthalten deshalb Anspielungen auf Monty Python (z. B. „spam“, „eggs“).

## Konzept: Interpreter

Python wird von einem **Interpreter** ausgeführt:

1. Du schreibst Quellcode (z. B. in einer `.py`-Datei oder in einer Notebook-Zelle).
2. Der Interpreter liest den Code, übersetzt ihn intern (u. a. in Bytecode) und führt ihn aus.
3. Es ist **kein separater Kompilerschritt** nötig wie bei C oder C++ – du startest das Programm direkt mit `python datei.py` oder, im Notebook, durch Ausführen der Zelle.

Vorteile für den Einstieg: schnelles Ausprobieren, sofortige Rückmeldung, keine Build-Umgebung.

## Konzept: Objektorientierung

In Python ist **alles ein Objekt**: Zahlen, Texte, Listen, Funktionen – alles hat einen Typ und kann über Attribute bzw. Methoden angesprochen werden. Programme können mit **Klassen** strukturiert werden: Eine Klasse ist der „Bauplan“, eine **Instanz** das konkrete Objekt.

Für den Einstieg reicht die **Idee**: Du musst noch keine eigenen Klassen schreiben. Wir nutzen zuerst einfache Datentypen (Zahlen, Strings, Listen) und Funktionen. Die Objektorientierung vertiefen wir später, wo sie für die KT-Inhalte nützlich ist.
