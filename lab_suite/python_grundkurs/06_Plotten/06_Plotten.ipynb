## Was ist Matplotlib?

**Matplotlib** ist eine weit verbreitete Bibliothek zum Erzeugen von **2D-Grafiken**: Kurven, Balken, Punkte, Histogramme. Du übergibst **Daten** (z. B. Listen oder NumPy-Arrays), wählst den **Plottyp** – und erhältst eine Figur, die du anzeigen oder speichern kannst. In Jupyter erscheint die Grafik oft direkt unter der Zelle.

## Installation und Import

Matplotlib ist **nicht** in der Python-Standardbibliothek. Installation z. B. mit:

```bash
pip install matplotlib
```

Im Programm importierst du üblicherweise das Untermodul **`pyplot`** unter dem Kürzel **`plt`**:

```python
import matplotlib.pyplot as plt
```

Nach dem Zeichnen rufst du **`plt.show()`** auf, um die Figur anzuzeigen. In Jupyter Notebooks wird `plt.show()` oft automatisch am Ende einer Zelle ausgeführt.

````python
import matplotlib.pyplot as plt

print("Matplotlib importiert. Version:", plt.matplotlib.__version__)
````

## Liniendiagramm / Zeitfunktion

Ein **Liniendiagramm** eignet sich, wenn du einen **Verlauf über eine Größe** (z. B. Zeit) darstellen willst – z. B. ein **Signal im Zeitbereich**: Abtastwerte über der Zeit. Du gibst die x-Werte (z. B. Zeitpunkte) und die y-Werte (z. B. Amplitude) an; `plt.plot(x, y)` verbindet die Punkte mit Linien.

````python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 1, 50)   # 50 Zeitpunkte von 0 bis 1
y = np.sin(2 * np.pi * 3 * t)  # Sinus, 3 Hz (Beispiel-Signal)

plt.figure(figsize=(8, 3))
plt.plot(t, y, color="blue")
plt.xlabel("Zeit (s)")
plt.ylabel("Amplitude")
plt.title("Signal im Zeitbereich (Sinus 3 Hz)")
plt.grid(True)
plt.show()
````

## X-Y-Plot (allgemeiner Zusammenhang)

Ein **X-Y-Plot** zeigt die Abhängigkeit **einer Größe von einer anderen** – nicht zwingend Zeit. Zwei Arrays gleicher Länge: x-Achse = erste Größe, y-Achse = zweite Größe. Ideal für Kennlinien, Verläufe, oder „y in Abhängigkeit von x“.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 30)
y = 0.5 * x ** 2 - 2 * x + 1   # Beispiel: Parabel

plt.figure(figsize=(6, 4))
plt.plot(x, y, "g-", linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("X-Y-Plot: y in Abhängigkeit von x")
plt.grid(True)
plt.show()
````

## Histogramm (Verteilung)

Ein **Histogramm** zeigt die **Verteilung** von Werten: Wie oft kommen welche Werte vor? Die x-Achse ist in **Klassen** (Bereiche) unterteilt, die y-Achse zeigt die **Häufigkeit** (oder relative Häufigkeit). Nützlich für Amplitudenverteilungen, Fehlerverteilungen oder Spektren (später).

````python
import matplotlib.pyplot as plt
import numpy as np

# Beispiel: 500 Zufallszahlen (normalverteilt)
daten = np.random.randn(500)

plt.figure(figsize=(6, 4))
plt.hist(daten, bins=25, color="steelblue", edgecolor="white")
plt.xlabel("Wert")
plt.ylabel("Häufigkeit")
plt.title("Histogramm: Verteilung der Werte")
plt.show()
````

## Scatterplot (Punktwolke)

Ein **Scatterplot** (Streudiagramm) zeigt **einzelne Punkte** (x, y) ohne Linien. Ideal, wenn du **Zusammenhänge** oder **Streuung** zwischen zwei Größen darstellen willst – z. B. Messpunkte, Korrelationen oder Kennfelder.

````python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
x = np.random.rand(50)
y = 2 * x + 0.5 + 0.2 * np.random.randn(50)  # ungefähr lineare Abhängigkeit + Rauschen

plt.figure(figsize=(6, 4))
plt.scatter(x, y, alpha=0.7, c="coral")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Scatterplot: Punktwolke (x, y)")
plt.grid(True)
plt.show()
````

## Kurz: Wann welcher Plot?

| Plottyp | Typischer Einsatz | Beispiel KT / Signale |
|---------|-------------------|------------------------|
| **Liniendiagramm** | Verlauf über Zeit oder eine Variable | Signal im Zeitbereich, Zeitfunktion |
| **X-Y-Plot** | y in Abhängigkeit von x (allgemein) | Kennlinie, Verlauf, Frequenzgang |
| **Histogramm** | Verteilung von Werten, Häufigkeiten | Amplitudenverteilung, Spektralanteile (später) |
| **Scatterplot** | Einzelpunkte, Zusammenhang zweier Größen | Messpunkte, Korrelation |

Mit diesen Grundlagen kannst du Listen und Arrays aus Berechnungen oder Messungen sofort **visuell** darstellen – und damit das Vorlesungskapitel **Signale und Spektren** sehr anschaulich unterstützen.
